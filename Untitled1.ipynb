```python
for img in glob.glob('masks/*'):
    im_name = img.split('/')[-1]
    im_name = im_name.split('_mask')[0]+'.npy'
    im = cv2.imread(img,0)
    im = process(im)
    if im_name.split('.')[0] in train:
        np.save(f"train_pre/y/{im_name}", im)
        #cv2.imwrite(f"train_pre/y/{im_name}", im)
    else:
        np.save(f"test_pre/y/{im_name}", im)
    del im
```

```python
def process(mask):
    mask = torch.nn.functional.one_hot(torch.tensor(mask).long(), num_classes=26)
    print(mask.shape)
    mask[:,:,1] = (mask[:,:,1] + mask[:,:,8] + mask[:,:,10] + mask[:,:,11])
    mask[:,:,1] = (mask[:,:,1] >= 1).long()
    cnt = list(range( mask.shape[2]))
    for i in [8,10,11]:
        cnt.remove(i)
    mask = mask[:,:,cnt]
    tmp = mask[:,:,-1].numpy()
    mask[:,:,-1] = mask[:,:,-2]
    mask[:,:,-2] = torch.tensor(tmp).long()
    return mask.numpy()
```
